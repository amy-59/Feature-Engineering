### 4. One hot Encoding with Top categories

```python
df['brand'].value_counts()
```

- As __Brand__ Column have lots of classes so we will set threshold 100 for it
- That is brands which have cars less than 100 we will store(club) it in one class.

```python
counts=df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold= 100
```

```python
counts[counts<threshold].index
#Basically we find cars brand <100 
```

```python
repl=counts[counts<threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```

- Here where we get the brand with <100 
- i.e. repl we will convert it store in uncommon column
